```python
import pandas as pd
```

```python
name2 = ['movie_ad','the name of the movie(year)','genre']   
movie = pd.read_table('movies.csv', sep='::', engine='python', names=name2)
                  
movie
```

```python
people = ['users_ad','gender','age','occupation','postal code'] 
users = pd.read_table('users.csv', sep='::', engine='python', names=people )
users
```

```python
rating2 = ['users_ad','movie_ad','rating','timestamp','sr']   
rating = pd.read_table('ratings.csv', sep='::', engine='python', names=rating2)
rating
```

```python
data = pd.merge(rating,users)
data
```

```python
data2 = pd.merge(data,movie)
data2
```

```python
men = data2.pivot_table('rating', index='the name of the movie(year)', columns=['gender'], aggfunc='mean',fill_value=0) 
men2=round(men)
men2
```

```python
men = data2.pivot_table('rating', index='the name of the movie(year)', columns=['age'], aggfunc='mean',fill_value=0) 
men3=round(men)
men3
```
